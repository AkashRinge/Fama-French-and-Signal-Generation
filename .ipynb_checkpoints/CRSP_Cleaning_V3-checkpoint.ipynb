```python
import pandas as pd
from pandas.tseries.offsets import MonthEnd
import numpy as np

# Read the CSV file into a DataFrame
csv_path = "C:\\ASAM_Local\\WRDS.CSRP.MSTOCK.ALL.9.27.23.csv"
mcrsp = pd.read_csv(csv_path, parse_dates=['date'], low_memory=False)

# Filter based on date range and share codes in a single step
mcrsp = mcrsp.loc[
    (mcrsp['date'] >= '1963-07-01') & 
    (mcrsp['date'] <= '2022-12-31') & 
    mcrsp['SHRCD'].isin([10, 11])
]

# Add new columns for month-end and year
mcrsp.loc[:, 'monthend'] = mcrsp['date'] + MonthEnd(0)
mcrsp.loc[:, 'year'] = mcrsp['date'].dt.year
# Rename SHROUT and calculate the market capitalization
mcrsp.rename(columns={'SHROUT': 'SHROUT_K'}, inplace=True)
# Add new columns for market capitalization

#CHANGE TO PRC!!!
mcrsp.loc[:, 'mktcap_M'] = abs(mcrsp["SHROUT_K"] * mcrsp["PRC"] / 10**6)

# Replace zero market caps with NaN
mcrsp.loc[mcrsp['mktcap_M'] == 0, 'mktcap_M'] = np.nan
```

```python
# Convert columns to numeric using NumPy
mcrsp['RET'] = pd.to_numeric(mcrsp['RET'], errors='coerce')
mcrsp['DLSTCD'] = pd.to_numeric(mcrsp['DLSTCD'], errors='coerce')
mcrsp['DLRET'] = pd.to_numeric(mcrsp['DLRET'], errors='coerce')

# Create NumPy arrays from DataFrame columns for faster access
RET = mcrsp['RET'].values
DLSTCD = mcrsp['DLSTCD'].values
DLRET = mcrsp['DLRET'].values

# Initialize an empty array for adjusted returns
ret_adj = np.empty_like(RET)
ret_adj.fill(np.nan)  # Fill with NaNs as default

# Define conditions for delisting using NumPy
cond1 = np.isnan(DLSTCD) | (~np.isin(DLSTCD, np.arange(500, 600)))
cond2 = ~np.isnan(DLSTCD) & ~np.isnan(DLRET) & np.isin(DLSTCD, np.arange(500, 600))

# Populate adjusted returns array based on conditions
ret_adj[cond1] = RET[cond1]
ret_adj[cond2] = DLRET[cond2]

# Assign the adjusted returns array back to the DataFrame
mcrsp['ret_adj'] = ret_adj

# convert the new column to numeric, although it should already be so
mcrsp['ret_adj'] = pd.to_numeric(mcrsp['ret_adj'], errors='coerce')
```

```python
# Convert columns to numeric using NumPy
mcrsp['SICCD'] = pd.to_numeric(mcrsp['SICCD'], errors='coerce')
mcrsp['EXCHCD'] = pd.to_numeric(mcrsp['EXCHCD'], errors='coerce')

# Create NumPy arrays for faster access
SICCD = mcrsp['SICCD'].values
EXCHCD = mcrsp['EXCHCD'].values

# Initialize empty arrays for industry and exchange
industry = np.empty(SICCD.shape, dtype=object)
exchange = np.empty(EXCHCD.shape, dtype=object)

# Define conditions for industry classification
industry[(1 <= SICCD) & (SICCD <= 999)] = "Agriculture"
industry[(1000 <= SICCD) & (SICCD <= 1499)] = "Mining"
industry[(1500 <= SICCD) & (SICCD <= 1799)] = "Construction"
industry[(2000 <= SICCD) & (SICCD <= 3999)] = "Manufacturing"
industry[(4000 <= SICCD) & (SICCD <= 4899)] = "Transportation"
industry[(4900 <= SICCD) & (SICCD <= 4999)] = "Utilities"
industry[(5000 <= SICCD) & (SICCD <= 5199)] = "Wholesale"
industry[(5200 <= SICCD) & (SICCD <= 5999)] = "Retail"
industry[(6000 <= SICCD) & (SICCD <= 6799)] = "Finance"
industry[(7000 <= SICCD) & (SICCD <= 8999)] = "Services"
industry[(9000 <= SICCD) & (SICCD <= 9999)] = "Public"
industry[SICCD > 9999] = "Missing"  # Default case

# Define conditions for exchange classification
exchange[np.isin(EXCHCD, [1, 31])] = "NYSE"
exchange[np.isin(EXCHCD, [2, 32])] = "AMEX"
exchange[np.isin(EXCHCD, [3, 33])] = "NASDAQ"
exchange[~np.isin(EXCHCD, [1, 31, 2, 32, 3, 33])] = "Other"  # Default case

# Assign back to DataFrame
mcrsp['industry'] = industry
mcrsp['exchange'] = exchange
```

```python
# Function to compute the compound return for a rolling period
def compute_return_for_period(series, period):
    return (1 + series).rolling(period).apply(lambda x: x.prod() - 1)

# Function to calculate the trailing returns for each stock over a period of months
def calculate_trailing_returns(df, months):
    df = df.copy()  # Create a copy to avoid the warning
    column_name = f'T{months}M_ret_adj'
    df[column_name] = df.groupby('PERMNO')['ret_adj'].transform(lambda x: compute_return_for_period(x, months))
    return df

# Assuming crsp2 is already defined and is a slice from another DataFrame
mcrsp = mcrsp.copy()  # Create a copy to avoid the warning

# Apply the function to calculate Trailing 12 Month (T12M) returns
mcrsp = calculate_trailing_returns(mcrsp, 12)
```

```python
# Function to compute the compound return for a rolling period
def compute_return_for_period(series, period):
    return (1 + series).rolling(period).apply(lambda x: x.prod() - 1)

# Function to calculate the trailing returns for RF over a period of months
def calculate_trailing_RF(df, months):
    df = df.copy()  # Create a copy to avoid the warning
    column_name = f'T{months}M_RF'
    df[column_name] = df.groupby('PERMNO')['RF'].transform(lambda x: compute_return_for_period(x, months))
    df[column_name].fillna(0, inplace=True)  # Fill NaN values with 0
    return df

# Assuming crsp2 is already defined and is a slice from another DataFrame
mcrsp = mcrsp.copy()  # Create a copy to avoid the warning

# Apply the function to calculate Trailing 12 Month (T12M) returns
mcrsp = calculate_trailing_RF(mcrsp, 12)
```

```python
from datetime import datetime

def rank_top_10_by_market_cap(df):
    # Convert 'date' to datetime if it's not already
    if df['date'].dtype != 'datetime64[ns]':
        df['date'] = pd.to_datetime(df['date'])
    
    # Extract year from 'date' and create a new column 'year'
    df['year'] = df['date'].dt.year
    
    # Filter data for the past 5 years
    current_year = datetime.now().year
    past_5_years_df = df[df['year'] >= (current_year - 5)]
    
    # Identify year-end records
    year_end_df = past_5_years_df[past_5_years_df['date'].dt.is_year_end]
    
    # Rank companies by 'mktcap_M'
    year_end_df['mktcap_rank'] = year_end_df.groupby('year')['mktcap_M'].rank(method='first', ascending=False)
    
    # Filter for top 10 companies by market cap for each year
    top_10_df = year_end_df[year_end_df['mktcap_rank'] <= 10]
    
    # Sort values for better readability
    top_10_df = top_10_df.sort_values(by=['year', 'mktcap_rank'])
    
    return top_10_df

# top_10_companies = rank_top_10_by_market_cap(mcrsp)
# print(top_10_companies)

top_10_companies = rank_top_10_by_market_cap(mcrsp)
print(top_10_companies)
```

```python
import matplotlib.pyplot as plt

# Group data by exchange and date, and count the number of securities
securities_per_exchange = mcrsp.groupby(['exchange', 'date']).size().reset_index(name='n')

# Create the plot
fig, ax = plt.subplots()

# Loop through each exchange to plot its data
for exchange in securities_per_exchange['exchange'].unique():
    subset = securities_per_exchange[securities_per_exchange['exchange'] == exchange]
    ax.plot(pd.to_datetime(subset['date']), subset['n'], label=exchange)

# Customize the plot
ax.set_title('Monthly number of securities by listing exchange')
ax.set_xlabel('Date')
ax.set_ylabel('Number of Securities')
ax.legend()

# Format the x-axis to show dates clearly
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Group by 'date' and 'exchange' to get the sum of market capitalizations
market_cap_per_exchange = mcrsp.groupby(['date', 'exchange']).apply(lambda group: pd.Series({
    'mktcap': group['mktcap_M'].sum()
})).reset_index()

# Sort within each group for plotting
market_cap_per_exchange.sort_values(['exchange', 'date'], inplace=True)

# Create the plot
fig, ax = plt.subplots()

# Loop through each exchange to plot its data
for exchange in market_cap_per_exchange['exchange'].unique():
    subset = market_cap_per_exchange[market_cap_per_exchange['exchange'] == exchange]
    ax.plot(pd.to_datetime(subset['date']), subset['mktcap'], label=exchange)

# Customize the plot
ax.set_title('Monthly market cap by listing exchange in billions (Not Inflation Adjusted)')
ax.set_xlabel('Date')
ax.set_ylabel('Market Cap in billions')
ax.legend()

# Format the x-axis to show dates clearly
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Group data by industry and date, and count the number of securities for each
securities_per_industry = mcrsp.groupby(['industry', 'date']).size().reset_index(name='n')

# Create a pivot table for better plotting in matplotlib
pivot_table = securities_per_industry.pivot(index='date', columns='industry', values='n').fillna(0)

fig, ax = plt.subplots(figsize=(14, 8))

for industry in pivot_table.columns:
    ax.plot(pivot_table.index, pivot_table[industry], label=industry)

# Formatting
ax.set_xlabel('Date')
ax.set_ylabel('Number of Securities')
ax.set_title('Monthly Number of Securities by Industry')
ax.legend(loc='upper left')
plt.grid(True)

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Group by date and industry, then sum the market cap for each group
market_cap_per_industry = mcrsp.groupby(['date', 'industry']).apply(
    lambda group: pd.Series({
      "mktcap": group['mktcap_M'].sum()
    })
).reset_index()

# Sort within each group for plotting
market_cap_per_industry.sort_values(['industry', 'date'], inplace=True)

# Create a pivot table for better plotting
pivot_table = market_cap_per_industry.pivot(index='date', columns='industry', values='mktcap').fillna(0)

fig, ax = plt.subplots(figsize=(14, 8))

# Loop through each industry to plot its market cap over time
for industry in pivot_table.columns:
    ax.plot(pd.to_datetime(pivot_table.index), pivot_table[industry], label=industry)

# Add formatting
ax.set_xlabel('Date')
ax.set_ylabel('Market Cap (in billions)')
ax.set_title('Monthly Market Cap by Industry (Not Inflation Adjusted)')
ax.legend(loc='upper left')
plt.grid(True)

# Show the plot
plt.show()
```
