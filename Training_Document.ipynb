# Task 2,3. Create Annual Return Datasets and Market Cap
- Efficient function for rolling product over a window = cumulative product of entire series / cumulative product shifted by window size
- Groupby ticker the dataset and calculate the cumulative return over 250 days for annual return series
- Market cap is shrout * prc

```python
def rolling_prod(series, window):
    """
    Compute rolling product over a window for a given series.
    """
    # Compute rolling product for entire series first
    rolling_cumprod = series.cumprod()

    # Compute rolling product over a given window by dividing the rolling cumprod at
    # the end of the window by the rolling cumprod at the beginning of the window
    return rolling_cumprod / rolling_cumprod.shift(window)

annual_returns = []
ticker_groups = returns_df.groupby('ticker')
returns_df['ret'] = pd.to_numeric(returns_df['ret'], errors='coerce')
for ticker, group in ticker_groups:
    df = group.copy()
    df['date'] = pd.to_datetime(df['date'])
    df['annual_ret'] = rolling_prod((1 + df['ret']), 250) - 1
    df['market_cap'] = df['shrout'] * df['prc'] * 1000
    annual_returns.append(df.dropna(subset=['annual_ret']))
    print(f"{ticker} ", end=' ')
annual_returns = pd.concat(annual_returns, ignore_index=True)
annual_returns
```

# Task 4. Risk factor exposure dataset

- Importing data from kenneth french library
- Download the dataset as specified in the guide
- Run a multivariate OLS regression on each ticker for each year on the data

```python
risk_factors = pd.read_csv('FF3_Factors.csv', skiprows=3)
unnamed_col = risk_factors.columns[0]
risk_factors['date'] = pd.to_datetime(risk_factors[unnamed_col], format='%Y%m%d')
risk_factors.drop(columns=[unnamed_col], inplace=True)
for col in risk_factors.columns:
    if col != 'date':
        risk_factors[col] = pd.to_numeric(risk_factors[col], errors='coerce')
        risk_factors[col] = risk_factors[col] * 0.01

risk_factors['Mkt-RF'].interpolate(method='linear', inplace=True)
risk_factors['SMB'].interpolate(method='linear', inplace=True)
risk_factors['HML'].interpolate(method='linear', inplace=True)
risk_factors['RMW'].interpolate(method='linear', inplace=True)
risk_factors['CMA'].interpolate(method='linear', inplace=True)
risk_factors['RF'].interpolate(method='linear', inplace=True)
risk_factors
```

```python
merged_df = pd.merge(annual_returns, risk_factors, how='left', on=['date'])
merged_df['date'] = pd.to_datetime(merged_df['date'])
merged_df
```

```python
ticker_groups = merged_df.groupby('ticker')
ff3_factor_list = []

for ticker, tick_group in ticker_groups:
    df = tick_group.copy()
    df['year'] = df['date'].dt.year
    year_groups = df.groupby('year')
    
    for year, group in year_groups:
        if len(group) > 10:
            group_copy = group.copy()
            group_copy['ret'].interpolate(method='linear', inplace=True)
            Y = group_copy['ret'] - group_copy['RF']
            X = group_copy[['Mkt-RF', 'SMB', 'HML', 'RMW', 'CMA']]
            X = sm.add_constant(X)
            model = sm.OLS(Y, X).fit()
            coeff = model.params
            current_tuple = (ticker, year, coeff[0], coeff[1], coeff[2], coeff[3], coeff[4], coeff[5], len(group_copy))
            ff3_factor_list.append(current_tuple)
    print(f"{ticker} ", end=' ')
    
ff3_data = pd.DataFrame(ff3_factor_list, columns=['ticker', 'yyyy', 'a', 'bm', 'bsmb', 'bhml', 'brmw', 'bcma', 'N'])
ff3_data
```
